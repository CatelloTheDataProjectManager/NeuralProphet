```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from neuralprophet import NeuralProphet
import warnings
warnings.filterwarnings("ignore")
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("AirPassengers.csv")
df.rename(columns = {'Month':'ds', '#Passengers':'y'}, inplace = True)
df['ds'] = pd.DatetimeIndex(df['ds'])
df.info()
```

```python
ax = df.set_index('ds').plot(figsize=(15, 12))
ax.set_ylabel('Monthly Number of Airline Passengers')
ax.set_xlabel('Date')

plt.show()
```

```python
m = NeuralProphet()
df_train, df_val = m.split_df(df, freq='M', valid_p = 0.2)
metrics = m.fit(df_train, freq='M', validation_df=df_val)
```

```python
metrics
```

```python
future = m.make_future_dataframe(df, periods=24, n_historic_predictions=len(df))
forecast = m.predict(future)
forecast
```

```python
forecast.info()
```

```python
fig = go.Figure()

fig.add_scatter( x=metrics.index, y=metrics['RMSE'], mode='lines', name="RMSE")
fig.add_scatter( x=metrics.index, y=metrics['MAE'], mode='lines', name="MAE")

fig.update_layout(
    
    yaxis_title='Error',
    title='Model',
    hovermode="x")

fig.show()
```

#### Visualize the forecast by model

```python
ax = forecast.set_index('ds').plot(figsize=(15, 12))
ax.set_ylabel('Monthly Number of Airline Passengers')
ax.set_xlabel('Date')

plt.show()
```

```python
from bokeh.plotting import figure, show

# Create a line plot with Bokeh
p = figure(x_axis_type='datetime', height=300, width=800, sizing_mode='stretch_both')

# Define custom colors for each line
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# Filter out None values and plot each line with custom colors
p.line(forecast.loc[forecast['y'].notnull(), 'ds'], forecast.loc[forecast['y'].notnull(), 'y'], legend_label='Actual data', line_width=2, alpha=0.8, color=colors[0])
p.line(forecast.loc[forecast['yhat1'].notnull(), 'ds'], forecast.loc[forecast['yhat1'].notnull(), 'yhat1'], legend_label='Forecast', line_width=2, alpha=0.8, color=colors[1])
p.line(forecast.loc[forecast['trend'].notnull(), 'ds'], forecast.loc[forecast['trend'].notnull(), 'trend'], legend_label='Trend', line_width=2, alpha=0.8, color=colors[2])
p.line(forecast.loc[forecast['season_yearly'].notnull(), 'ds'], forecast.loc[forecast['season_yearly'].notnull(), 'season_yearly'], legend_label='Seasonal component', line_width=2, alpha=0.8, color=colors[3])

# Add axis labels and title
p.xaxis.axis_label = 'Date'
p.yaxis.axis_label = 'Monthly number of airline passengers'
p.title.text = 'Forecast of monthly airline passengers'

# Show the legend
p.legend.location = 'top_left'

# Display the plot
show(p)
```

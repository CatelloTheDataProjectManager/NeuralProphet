```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from neuralprophet import NeuralProphet
import warnings
warnings.filterwarnings("ignore")
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("AirPassengers.csv")
df.rename(columns = {'Month':'ds', '#Passengers':'y'}, inplace = True)
df['ds'] = pd.DatetimeIndex(df['ds'])
df.info()
```

```python
ax = df.set_index('ds').plot(figsize=(15, 12))
ax.set_ylabel('Monthly Number of Airline Passengers')
ax.set_xlabel('Date')

plt.show()
```

```python
m = NeuralProphet()
df_train, df_val = m.split_df(df, freq='M', valid_p = 0.2)
metrics = m.fit(df_train, freq='M', validation_df=df_val)
```

```python
metrics
```

```python
future = m.make_future_dataframe(df, periods=24, n_historic_predictions=len(df))
forecast = m.predict(future)
forecast
```

```python
fig = go.Figure()

fig.add_scatter( x=metrics.index, y=metrics['RMSE'], mode='lines', name="RMSE")
fig.add_scatter( x=metrics.index, y=metrics['MAE'], mode='lines', name="MAE")

fig.update_layout(
    
    yaxis_title='Error',
    title='Model',
    hovermode="x")

fig.show()
```

```python
import pytz
local_tz = pytz.timezone('America/New_York')
df['ds'] = df['ds'].dt.tz_localize(local_tz)
forecast['ds'] = forecast['ds'].dt.tz_localize(local_tz)
```

#### Visualize the forecast by model

```python
fig = go.Figure()

fig.add_scatter(x=df['ds'], y=df['y'], mode='lines', name="original")
fig.add_scatter(x=forecast['ds'], y=forecast['yhat1'], mode='lines', name="prediction")

# Assuming 'trend' and 'season_yearly' are columns in the 'forecast' dataframe
fig.add_scatter(x=forecast['ds'], y=forecast['trend'], mode='lines', name="trend")
fig.add_scatter(x=forecast['ds'], y=forecast['season_yearly'], mode='lines', name="season_yearly")

fig.update_layout(
    yaxis_title='Passengers',
    title='Forecasts',
    hovermode="x")

fig.show()
```
